```python
# Travel along the x-axis and print each value
def recursive_axis_X(matrixA2, rowY,  colX, final_X):
     if colX == final_X: # value seeds

            return 0 # neutral value
     else:
         print(matrixA2[rowY][colX], "   ", end=' ')
          # Recursive sum x
         return 1 + recursive_axis_X(matrixA2, rowY, colX + 1, final_X)

#embedded row and column
def recursive_axis_Y(matrixA2, rowY, final_Y):
    if rowY == final_Y: # value seeds
        return 0 # neutral value
    else:
        print("\n")
        v = recursive_axis_X(matrixA2, rowY, 0, final_Y)
        # Recursive sum y
        return 1 + recursive_axis_Y(matrixA2, rowY + 1, final_Y)
# ----------------------------------------------------------------
# Run axis x from diagonal's value
def recursive_axis_X1(colX,  finalX):
    if colX == finalX: # value seeds, final value
            return 0 # neutral value
    else:
           print(colX , " ", )
          # Recursive sum x
           return 1 + recursive_axis_X1(colX+1, finalX)

# Run axis x from diagonal's value
def recursive_axis_X2(matrixA2,diagonalPivot, rowY, colX,  finalX):
    if colX == finalX: # value seeds, final value
            return 0 # neutral value
    else:
           v1 = matrixA2[diagonalPivot][diagonalPivot]
           v2 = matrixA2[diagonalPivot][colX]

           f1 = matrixA2[rowY][diagonalPivot]
           f2 = matrixA2[rowY][colX]

           n =  (v1 * f2) - (f1 * v2)
           matrixA2[rowY][colX] = n

           print(v1," * " ,f2," - " , f1 , " * ", v2 )
          # Recursive sum column x
           return 1 + recursive_axis_X2(matrixA2, diagonalPivot, rowY, colX+1, finalX)

# Run each row Y2 after pivot
def recursive_row_Y2(matrixA2, diagonal,  rowY, final_Y):
    if rowY == final_Y: # value seeds
        return 0 # neutral value
    else:

         print( diagonal, " Row y2 ", rowY)
         recursive_axis_X2(matrixA2,diagonal, rowY ,diagonal ,  final_Y)

        # Recursive sum
         return 1 + recursive_row_Y2(matrixA2, diagonal,  rowY + 1, final_Y)

# Run diagonal y1 - pivot
def recursive_Print_Lower_Diagonal(matrixA2, diagonal, final_Y):
    if diagonal == final_Y: # value seeds
        return 0 # neutral value
    else:
         print(" Diagonal",diagonal, "\n")
         recursive_row_Y2(matrixA2,  diagonal, diagonal + 1, final_Y)
         #print("," "\n")
        # Recursive sum
         return 1 + recursive_Print_Lower_Diagonal(matrixA2, diagonal + 1, final_Y)

def main():
# Sum diagonal recursive
    matrixA = [[1,3,2,1,9],
               [2,7,2,3,7],
               [1,4,5,2,3],
               [3,2,1,1,2],
               [8,7,6,3,4]]

    vector1 = [0]
    vector2 = [0]

    rowA = len(matrixA)

    s = recursive_axis_Y(matrixA, 0, rowA)
    print("\n")
    print("-----6", end=' ')
    print("\n")
    recursive_Print_Lower_Diagonal(matrixA, 0, rowA)
    print("\n")
    s = recursive_axis_Y(matrixA, 0, rowA)

main()
```
